```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('musteriler.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cinsiyet = le.fit_transform(df['Cinsiyet'].values)
cinsiyet = pd.DataFrame(cinsiyet)
X = pd.concat([cinsiyet,df.iloc[:,2:]], axis=1)
X.rename(columns={0:'cinsiyet'}, inplace=True)
X.head()
```

```python
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(linkage='ward', n_clusters=2, affinity='euclidean')
ac_pred = ac.fit_predict(X)
ac_pred
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x=X.iloc[ac_pred == 0,2], y=X.iloc[ac_pred==0,3], s=100, c='orange')
plt.scatter(x=X.iloc[ac_pred == 1,2], y=X.iloc[ac_pred==1,3], s=100, c='black')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(linkage='ward', n_clusters=3, affinity='euclidean')
ac_pred = ac.fit_predict(X)
ac_pred
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x=X.iloc[ac_pred == 0,2], y=X.iloc[ac_pred==0,3], s=100, c='orange')
plt.scatter(x=X.iloc[ac_pred == 1,2], y=X.iloc[ac_pred==1,3], s=100, c='black')
plt.scatter(x=X.iloc[ac_pred == 2,2], y=X.iloc[ac_pred==2,3], s=100, c='red')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
ac = AgglomerativeClustering(linkage='ward', n_clusters=4, affinity='euclidean')
ac_pred = ac.fit_predict(X)
ac_pred
```

```python
plt.figure(figsize=(16,9))
plt.scatter(x=X.iloc[ac_pred == 0,2], y=X.iloc[ac_pred==0,3], s=100, c='orange')
plt.scatter(x=X.iloc[ac_pred == 1,2], y=X.iloc[ac_pred==1,3], s=100, c='black')
plt.scatter(x=X.iloc[ac_pred == 2,2], y=X.iloc[ac_pred==2,3], s=100, c='red')
plt.scatter(x=X.iloc[ac_pred == 3,2], y=X.iloc[ac_pred==3,3], s=100, c='blue')
plt.show()
```

```python
import scipy.cluster.hierarchy as sch 
plt.figure(figsize=(16,12))
dendogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.show()
```
